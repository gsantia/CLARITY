# Response time EDA

```python
import main, threads, users
import datetime, json, numpy, math
from matplotlib import pyplot as plt
from collections import defaultdict
%matplotlib inline
```

## All Users
First we want to look at general statistics regarding response times across the board.
So load the user data:

```python
#store the response times in dicts for quick access
all_dict = {user : main.userResponse(user) for user in main.COMMENTS_BY_USER}
top_dict = {user : main.userResponse(user, "Top") for user in main.COMMENTS_BY_USER}
reply_dict = {user : main.userResponse(user, "Reply") for user in main.COMMENTS_BY_USER}

all_comments = []
user_averages = []
for user in all_dict:
    times = all_dict[user]
    all_comments.extend(times)
    user_averages.append(numpy.mean(times))


    
```

### total comments


```python
fig = plt.figure(figsize = (8,8))
plt.hist(all_comments)
#find the average overall
total_average = numpy.mean(all_comments)
print total_average
```

```python
all_replies = []
for user in reply_dict:
    all_replies.extend(reply_dict[user])
fig = plt.figure(figsize = (8,8))
plt.hist(all_replies)

#find the average
print numpy.mean(all_replies)
    
```

### top-level

```python
all_top = []
for user in top_dict:
    all_top.extend(top_dict[user])
fig = plt.figure(figsize = (8,8))
plt.hist(all_top)

#find the average
print numpy.mean(all_top)

    
```
